```python
# Load Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
#Flattening each input image
X_train_flat = X_train.reshape(X_train.shape[0], 28*28)
X_test_flat = X_test.reshape(X_test.shape[0], 28*28)

#Normalizing
X_train_normalized = X_train_flat/255
X_test_normalized = X_test_flat/255
```

```python
from joblib import load
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score, ShuffleSplit
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
import numpy as np
import pandas as pd

def train_classifiers(estimator, X_train, y_train, cv, name):
    estimator.fit(X_train, y_train)
    cv_train_score = cross_val_score(estimator, X_train, y_train, cv=cv, scoring='f1_macro')
    print(f"On average, {name} model has an f1 score of"
          f" {cv_train_score.mean(): .3f} +/- {cv_train_score.std():.3f} on the training set")

def eval(estimator, X_test, y_test):
    y_pred = estimator.predict(X_test)
    
    # Print precision, recall, and F1 score for each class
    class_report = classification_report(y_test, y_pred)
    print("Classification report:\n", class_report)
    
    # Print confusion matrix
    print("\nConfusion matrix:")
    cm = confusion_matrix(y_test, y_pred)
    accuracy_test = accuracy_score(y_test, y_pred)
    precision_test = precision_score(y_test, y_pred, average='weighted')
    recall_test = recall_score(y_test, y_pred, average='weighted')
    f1_test = f1_score(y_test, y_pred, average='weighted')
    # Print evaluation metrics for the test set
    print(f"Test Accuracy: {accuracy_test * 100:.2f}%")
    print(f"Test Precision: {precision_test:.4f}")
    print(f"Test Recall: {recall_test:.4f}")
    print(f"Test F1-Score: {f1_test:.4f}")
    print("\n" + "=" * 40 + "\n")
    print("Confusion Matrix:")
    
    # Display confusion matrix plot
    x = ConfusionMatrixDisplay(confusion_matrix=cm)
    x.plot()
```

```python
# Load the pre-trained models
random_forest_pipeline = load('random_forest_pipeline.joblib')
random_forest_pipeline_hypertuned = load('random_forest_pipeline_hypertuned.joblib')

# Evaluate the basic random forest pipeline
print("Evaluation for Basic Random Forest Pipeline:")
eval(random_forest_pipeline, X_test_normalized, y_test)

# Evaluate the hyper-tuned random forest pipeline
print("Evaluation for Hyper-tuned Random Forest Pipeline:")
eval(random_forest_pipeline_hypertuned, X_test_normalized, y_test)
```
